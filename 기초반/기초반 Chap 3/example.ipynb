# 기초반 3차시 예제 코드

## 1. Condition Statement(조건문)

### 1-1. if

#### if문의 내용이 참일 경우

```python
if 102958372 % 2 == 0:
    print("This is even number.")
```

#### if문의 내용이 거짓일 경우

```python
if 102958371 % 2 == 0:
    print("This is even number.")
print("This is odd number.")
```

### 1-2. if-else

```python
if 102958371 % 2 == 0:
    print("This is even number.")
else:
    print("This is odd number.")
```

### 1-3. if-elif-else

```python
if 102958371 % 2 == 0:
    print("This is even number.")
elif 102958371 % 2 == 1:
    print("This is odd number.")
else: print("Is this 0?")
```

### 1-4. (match-case문)

```python
match 102958371 % 2:
    case 0:
        print("This is even number.")
    case 1:
        print("This is odd number.")
    case _:
        print("Is there something wrong?")
```

## 2. Loop(반복문)

### 2-1. for

#### list에 대한 for문

```python
for country in ["Italy", "France", "Germany", "England"]:
    print(country)
```

#### tuple에 대한 for문

```python
for country in ("Italy", "France", "Germany", "England"):
    print(country)
```

#### range에 대한 for문

```python
for i in range(5):
    print(i)
```

### 2-2. while

```python
cnt = 0
while cnt < 5:
    print(cnt)
    cnt += 1
```

#### while문을 통한 무한루프

```python
while True:
    print(1)
```

### 2-3. break, else, pass, continue

#### break

```python
for i in range(5):
    if i == 4:
        print("i == 4 and break")
        break
    else: print(i)
```

#### else

```python
for i in range(5):
    print(i)
else:
    print("This is else code block.")
```

#### break-else 복합문

```python
for i in range(5):
    if i == 4:
        print("break is occured.")
        break
else:
    print("break is not occured.")
```

```python
for i in range(5):
    if i == 5:
        print("break is occured.")
        break
else:
    print("break is not occured.")
```

#### continue

```python
for i in range(10):
    if i % 2 == 0:
        print(i)
    else: continue
```
